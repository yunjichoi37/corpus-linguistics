```python
import os, re
from collections import Counter
import pandas as pd
os.mkdir('./Q1_Out')
```

```python
# 1번 답
for f_name in os.listdir('./Q1'):
    final = [j for i in open('./Q1/' + f_name, 'r').readlines() for j in re.findall('<[wcmw]+.+?">(.+?)\s?</[wcmw]+>', i)]
    open('./Q1_Out/' + f_name[:-4] + '.out', 'w', encoding='utf-8').write('\n'.join(final))
```

```python
# diff -u --strip-trailing-cr Q1_Out Q2
```

```python
# 2번 답
TDM = {i[:-4]: Counter(j.lower()
                for j in open('./Q2/' + i).read().split()
                if re.search('[aeiou]{2,}', j.lower()))
        for i in os.listdir('./Q2')}

df = pd.DataFrame(TDM).sort_index().fillna(0).astype('int')
df.to_csv('./Q2_Out.txt', sep='\t')
```
