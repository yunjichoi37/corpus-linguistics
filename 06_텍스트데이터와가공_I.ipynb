## 연습문제 3

1. `[06_out]` 폴더 만들기
2. `[06_data]` 폴더 파일 중, .txt 파일만 불러오기
3. 각 파일의 텍스트를 어휘 단위로 분리하기(공백 문자 연쇄 기준으로).
4. 어휘 앞뒤의 문장부호 하나 이상 연쇄 제거하기
5. 첫 문자가 대문자인 어휘만 추출하여 오름차순으로 정렬하기
6. 확장자를 out으로 변경하여, 5의 처리 결과를 하나의 라인에 하나의 어휘로 `[06_out]` 폴더에 출력하기

```python
os.mkdir('06_data/06_out')
```

```python
txts = [i for i in os.listdir('06_data') if i.endswith('.txt')]; txts
```

```python
words = list()
num = 0
for txt in txts:
    words.append(open('06_data/' + txt).read().split())

    final = [i.strip('.,"\'') for i in words[num] if i.istitle()]
    final.sort()
    num += 1
    open('06_data/06_out/0' + str(num) + '.out', 'w').write('\n'.join(final))
```
