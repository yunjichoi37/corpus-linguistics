## Markup language
- SGML > HTML > XML
- 문서가 화면에 표시되는 형식, 데이터의 논리적인 구조 등을 정의한 언어이다.
<br><br>
- HTML: 형식적인 부분만 각괄호<>로 표시

## SGML/HTML/XML 간의 비교

- SGML
  - 사용자마다 개별로 정의한 마크업 태그를 사용할 수 있다.
  - 공부하기 어려움
- HTML
  - 형식적으로 지정된 마크업만 사용할 수 있다.
  - 공부하기 쉬움
  - 눈에 보이는 형식적 태그만 정의되어 있다는 게 단점이다.
  - 따라서 의미 표현이 어렵고, DB로 정의하기 어렵다.
- XML
  - 위 둘의 난이도를 중간 정도로 조정함.
  - 의미가 있는 태그를 활용할 수 있다.
  - 의미 표현을 나타낼 수 있기에 DB로 활용가능성을 높일 수 있다.

## 리스트 내장 list comprehension
사용 형식<br>
[출력식 for 변수명1 in 객체1(if1)<br>
for 변수명2 in 객체2(if2)]

```python
[k * k for k in range(10)
        for j in range(4) if j % 2==0]
```

```python
L = list()
for k in range(10):
    L.append(k * k)
L
```

## 데이터 구조와 내장

```python
(i for i in range(5)) # generator
```

```python
tuple(i for i in range(5)) # tuple
```

```python
{i for i in range(5)} # set
```

```python
{i: i * i for i in range(5)} # dictionary
```

## 소요시간 측정
- %: line command
- %%: cell command

```python
%%timeit
[k for k in range(100000000)]
```

```python
%%timeit
L = []
for k in range(100000000):
    L.append(k)
```

```python
%%timeit
L = []
t = range(100000000)
for k in t:
    L.append(t[k])
```

## 연습문제 3을 리스트 내장 코드로

```python
b = re.findall(r'\d+\t.+\t(.+)\n', a)

c = [j for i in b 
        for j in re.split(r'\s[+]\s', i)]
```

## 빈도 추출
- collections의 Counter는 독자적인 구조를 가지고 있다.
  - collections.Counter
  - {}을 품은 Counter
- `most_common(n)`
  - 고빈도 n개를 [(),(), ..., ()]로 반환한다.
  - Counter type에만 동작한다.

```python
import collections
d = collections.Counter(c)
type(d)
```

```python
d
```

```python
d.most_common(10)
```

```python
d.items() # 빈도순으로 정렬하지 않고 원래의 순서대로 출력된다.
```

## 구조 변환: Counter > Series

```python
import pandas as pd
e = pd.Series(d)
e.head()
```

```python
e.sort_values(ascending=False).head()
```

```python
## dataframe으로 바꾸는 방법!!
e = pd.DataFrame({'Freq':d}); e
```
