# Corpus

## 통계학과 표본
- 1895년 경 모집단의 축소판인 표본조사가 제안되었다. 
- sample = miniature population
- 표본의 대표성에 대한 개념이 제안되었다.
<br><br>

- 1934년: 무작위 표본 추출(random sampling)이 중요하다.
    - 최소 크기는 대체로 1천

## 기본적인 표본 추출 방법
1. 우연성
- 단순 무작위 표본 추출: 난수, 제비뽑기
- 체계적 표본 추출: 길거리 설문조사 + 일정간격 추출
- 개체가 표본에 **포함될 동등한 기회**를 제공한다.
<br><br>
2. 분포 고려: **층화 표본 추출(startified sampling)**
- 모집단을 구성하는 **층화/집단의 분포 비율을 고려**해서 추출한다.
  - **층화 내에서는 우연성을 통해 표본 추출**
- 여론조사, 코퍼스

## Corpus
- 연구 대상이 될 수 있는 전산화된 문자 텍스트, 음성 자료의 집합
- 일반적 의미: 언어 연구 목적으로 대표성, 균형성, 크기를 고려하여 설계됨.
- 광의적 의미: 반드시 대표성/균형성이 고려될 필요는 없다. **대규모**라면 모집단의 특성을 알 수 있지 않을까?
- 언어학적 가치
  - 모국어 화자의 언어 특성이 반영된 대규모 언어 데이터
  - 인위적으로 생성되지 않음 > 의사소통 맥락 반영
  - 직관적 판단 어려움 > 마음과 표현 사이 상관관계

## 대표성
- 표본: 모집단의 특성을 대표하는 것
  - 표본의 크기, 균형성 등 고려
- 코퍼스
  - 코퍼스 설계: 변이성 고려, 대표성(균형성/크기) 고려
  - 층화표본추출: 층화 내에서 random sampling
    - 여론조사
    - Brown Corpus: 장르별 출판 비율을 고려하여 코퍼스를 구성함.
  - 모집단을 대표하는 표본으로서의 코퍼스 >> 보편성/변이성 관찰 가능
  - 모집단을 비교하기 위한 코퍼스: 참조 코퍼스 > 개별 텍스트와 비교

## Brown Corpus
- 세계 최초로 구축: 코퍼스의 전형, 효용성 입증됨.
- 표본 추출 방법: 당시 미국의 출판 비율을 고려하여 구성함.
  - 장르 별 텍스트에 일련 번호를 부여한 후 제비뽑기
  - 층화표본추출 + 단순 무작위 표본 추출
- LOB Corpus: 영국영어

## BNC(British National Corpus)
- 문어 90%, 구어 10%
- 대표성을 위한 설계 기준
  - 장르 다양성 유지
  - 표본이 4천 어휘 초과 X
  - 화자, 작가의 변수 고려
  - 문어의 경우 출처를 정확히 기술
  - 방언을 대표할 수 있도록 인구 통계 고려
  - 사회 계층 분포 고려

## 21세기 세종계획 코퍼스
- 한국어를 진작시키기 위한 과제
- 한영/한일 병렬 코퍼스, 북한/해외 한국어 코퍼스, 역사 자료, 전문 용어
- 현대국어 기초: 원시, 형태분석, 형태의미분석, 구분문석
- 현대국어 구어 전사: 원시, 형태분석

## 코퍼스의 다양한 종류
- 언어적 측면: 외국어학습자, 언어습득, 언어장애환자, 방언, 음성, 감성분석, ...
- 국립국어원 모두의 말뭉치 for 인공지능 개발
- 역사문화적 측면

## 빅데이터의 한계
- 규모보다 용도에 맞는 데이터가 더 중요하다.
  - 새로운 알고리즘보다, 적용 가능한 제대로 된 알고리즘이 중요하다.
- 트위터: 인구통계적 특성이 일반적이라 할 수 있으므로 대표 특성으로 다루는 것은 조심해야 한다.
- 데이터를 분류하고 판단하는 것 또한 사람이라서 편향될 수 있다.
- 블랙박스 모델이기 때문에 답변이 어떻게 나왔는지 알 수 없다.
- 언어적 다양성을 반영하지 못하고 영어의 지배력을 키울 수 있다.

# 빅데이터와 융합
## 빅데이터의 본질과 융합
- 빅데이터 활용: 고객의 행동을 예측하고 대처 방안을 마련하여 기업의 경쟁력 강화, 생산성 향상과 비지니스 혁신
- 빅데이터 활용 목적: 인간과 세상에 대한 이해로 예측과 문제 해결
- 역사 기록의 성격이 바뀌면서 과학과 인문학 사이 경계가 뒤섞이고 있다. '디지털 인문학자', '컴퓨터사회과학'

## 데이터 과학과 융합
- 데이터 수집 > 가공 > 분석/시각화 > 스토리텔링
- data mining > data science
- 숨겨져 있던 의미를 찾아서
- 데이터 과학 인재의 특성: 기술 전문성, 호기심, 스토리텔링 능력, 창의적인 관점
- 반짝이는 눈을 가진 탐험가 콜롬버스와 의심 많은 형사 콜롬보

## 코퍼스 언어학과 텍스트 마이닝
- Corpus linguistics: 전산 언어 자원을 구축하고, 이것을 바탕으로 컴퓨터를 이용하여 언어 연구와 응용을 수행한다.
- Text mining: 상용적 목적에서 텍스트의 패턴/관계/의미를 분석하고 통찰력을 발굴하고자 함이다.
- 지향점
  - 코퍼스: 개인 연구자의 관찰 한계를 넘는 규모, 변이성과 보편성이 반영된 데이터
  - 통계: 관찰할 수 없던 것을 관찰할 수 있는 기회를 제공하는 현미경같은 도구
  - 직관으로 포착하기 어렵던 것을 포착

## 기술만으로 데이터를 하는 것은 아니다
- 인간을 심도있게 연구하려는 분야, 인문학적 통찰을 가지고 가설을 세우고 연구방식
- 인문학 + 기술 = 더 높은 수준 협업 가능할 것 같다.
- 담론의 전체 지형을 그리고, 움직임을 관찰하고, 의미를 파악하기

## 21세기 문헌학으로서 코퍼스언어학과 텍스트 마이닝
- 문헌학(philology): 역사적 문헌과 문자를 연구한다.
  - 역사적 원형과 의미에 대한 연구를 진행한다.
- 빅데이터 환경: 현재의 기록으로부터 미래 예측
- 빅데이터 활용의 지향점은 대개 인간이다.
- 학제간 연구가 필요하다. 여러 학문에 대한 관심과 공감 능력을 요한다.

1. Brown 코퍼스에서 사용한 표본 추출 과정을 구체적으로 제시하시오.<br>
: 당시 미국의 출판 비율을 고려하여 장르 별로 텍스트 비율을 구성하였다. <br>
장르 별 후보 텍스트들에 번호를 부여하여 제비뽑기를 통해 코퍼스를 구축할 표본 텍스트들을 선정하였다.
2. Brown 코퍼스 구축에서 사용된 표본 추출 방법을 통계학에서는 무엇이라고 하는가?<br>
: 장르 별 텍스트 비율 구성은 층화 표본 추출, 제비뽑기는 단순 무작위 표본 추출이라 한다.
3. Brown 코퍼스 구축에서 이러한 통계적 표본 추출 방법을 활용하는 이유는 무엇인가?<br>
: 우선, 모집단 전체를 전수 조사하기 어렵다는 근본적인 문제가 있다.<br>
그 후에는 모집단의 대표성과 균형성을 반영하기 위해 장르 분포 비율을 고려하였고, 동일 장르 내에서는 랜덤으로 고려하였다.

<빅데이터 시대에서 왜 텍스트 데이터가 중요한가>
1. 기업의 빅데이터 활용 목적은 무엇인가?<br>
: 기업에서 빅데이터를 활용하고자 함은 고객의 행동을 예측하고, 그에 대한 대처 방안을 마련하기 위함이다. 이렇게 해서 기업의 경쟁력을 마련하면, 생산성 향상과 비지니스 혁신을 이룰 수 있다. 궁극적으로는 인간과 세상에 대한 이해도 담겨있을 것이다.
2. 1에서 제시한 활용 목적에 부합하는 텍스트 데이터의 활용 가치는 무엇인가?
: 직관적인 판단의 한계를 극복할 수 있다. 코퍼스 언어학과 통계를 통해 개인 연구자가 알 수 없는 여러 패턴과 의미를 알아낼 수 있다. 이를 통해 고객의 반응과 예측을 텍스트 데이터에서 추출하여 활용할 수 있다.
