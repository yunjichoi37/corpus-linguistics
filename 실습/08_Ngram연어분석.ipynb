# N-Gram 및 연어 이해
## 빈도 분석의 한계
- 빈도 목록
    - 타입(어휘/형태소) ~ 토큰 수(빈도 수) 대응쌍
- 타입
    - 타입은 어휘/형태소 등 형태적 동일성 기반 언어 단위이다.
    - 맥락 제거 > 의미 파악 한계
    - 중의성(ambiguity)
    - 모호성(vagueness)
- 맥락을 살펴볼 수 있는 분석
    - 용례분석(concordance)
    - N-gram
    - 연어분석(collocation)

## N-gram
- uni-gram: 하나의 어휘가 하나의 분석 단위
- bi-gram: 두 어휘 연쇄가 하나의 분석 단위
- tri-gram: 세 어휘 연쇄가 하나의 분석 단위
- four-gram, five-gram, ...

## 친숙성 효과 familiarity effect
- 사용 빈도가 높은 어휘일수록, 즉, 친숙한 어휘일수록 뇌에서 더 빨리 처리
- 어휘 판단(lexical decision) 실험
    - 언어음/문자열 연쇄 제시 후 어휘인지 아닌지 판단하는 실험
    - 반응 시간(response time) 측정 $\Rightarrow$ 반응 시간이 오래 걸릴수록 다른 인지적 과제 수행 가정
    - **사용 빈도가 높을수록 반응 시간이 빠르다.**

## 촉발 효과 priming effect
- 어휘 판단 실험에서 어떤 어휘 제시 후 형식적 또는 의미적으로 관련된 어휘를 제시하면 반응 시간이 빨라진다.
- nurse 제시 후 ⇒ 의미적으로 연관되지 않은 flower보다 의미적으로 관련된 doctor의 반응 시간이 빠르다.
- 형식적 촉발 효과
    - mouse는 쥐, 컴퓨터 마우스 중의성(동음이의어)
    - 피험자에게 The mouse ate the cheese. 제시
    - 문장에서 mouse는 쥐의 의미로 해석
    - cup보다 의미적 연관성이 높은 cat의 반응 시간이 빠르다.
    - 위 문장에서 mouse는 의미적으로 컴퓨터 마우스와 관련이 없지만 형식적으로 관련이 되어 computer의 반응 시간이 빨라진다.

## 분포 가설 Distributional Hypothesis
- Harris, Z. (1954): "Words that are used and occur in the same contexts tend to purport similar meanings."
- Firth, J.R. (1957): "A word is characterized by the company it keeps."
- 인지과학, 전산학 등에서 어휘의 의미와 관계를 접근하는 토대
    - 단어 임베딩(Word Embedding) 기술 기반
- 함께 자주 나타나는 단어들은 유사한 의미를 가질 가능성/관련될 가능성이 높다.

## 언어 collocation/collocates
- 친숙성/빈도 효과, 촉발 효과 $\Rightarrow$ 동일 문장/문단/맥락에서 빈번하게 함께 사용되는 두 어휘쌍
- 영어 화자의 경우
    - cow와 milk가 빈번하게 함께 사용
    - 연상 실험에서 cow를 자극어로 제시하면 많은 사람들이 milk를 반응어로 반응
- 연어의 활용
    - 머릿속 사전(mental lexicon)에서 어휘를 어떻게 저장하고 검색하는가? $\Rightarrow$ 심리언어학, 인지과학, 신경과학 등
    - 어휘 의미를 어떻게 습득하고, 이해하는가? 다른 어휘 의미와 어떻게 식별되는가? $\Rightarrow$ 의미론, 화용론, 언어교육 등
    - 텍스트의 특정 주제는 어떤 어휘군으로 구성되는가? $\Rightarrow$ 텍스트 마이닝, 문서 분류, 문서 요약 등
    - 사회와 문화의 특성은 어떤 어휘군으로 표상되는가? $\Rightarrow$ 인문사회과학
- 여러 문서에 빈번하게 함께 나타나는 단어들을 추출해보기

#### Kennedy, Graeme. (2003). “Amplifier collocations in the British National Corpus: Implications for English language teaching”. Tesol Quarterly 37-3, 467–87.

- amplifier(확장사)
    - absolutely, completely, really, very
    - 수식받는 형용사 등에 대해 가정된 기준보다 강조의 정도를 확장
    - Maximizer
        - absolutely thrilled, completely unclear, totally devastated, utterly ruthless.
        - 형용사, 동사의 의미를 최대로 강화
    - Booster
        - really annoyed, particularly helpful, highly skilled, incredibly stupid, deeply suspicious, enormously grateful, severely depressed
        - Maximizer보다 강화 정도가 약한

- amplifier collocation: 선행 연구
    - Sinclair (1991)
        - **코퍼스 기반 접근하여 언어 교육 및 언어 사용에서 연어의 중요성 제시**
        - thank you very much vs. ?? thank you much
    - Lorenz(1999)
        - L2 학습자는 특정 부사만 사용하는 경향
    - Biber et al.(1999)
        - British English 구어 높은 빈도: very good, very nice, really good, really nice, too bad
        - American English 구어 높은 빈도: really good, too bad, very good, real good, real quick, really bad, really nice, too big, very nice
        - 외국어 학습자는 이러한 측면을 알아채기 어렵다.
        - 화자와 작가마다 형용사를 수식하는 정도 부사 사용에 다양성을 보인다.
        - 대부분의 경우 completely different, totally different와 같이 의미 변화 없이 또는 의미 변화가 거의 없이 교체 가능 $\Rightarrow$ **Kennedy(2003)은 이러한 측면에 차이가 있지 않을까에 대한 접근**

- Absolutely
    - 과장(hyperbolic) 의미 형용사와 공기(co-occurrence) 경향: fabulous, marvellous, fantastic
    - 부정 의미의 접두사는 incredible이 유일하다.
    - -ous 접미사(23%), -ed 접미사 (15%)

- Completely
    - elimainated(제거된, 탈락된), eradicated(근절된) 등 제거 의미 형용사
    - 부정 의미 접두사 23%
    - out-/over- 접두사 10%, -ed 접미사 78%

## zip 함수
- 직접 N-gram을 만들어보자.
- 출력 결과가 zip type
- iterable object를 하나의 튜플로 연결시켜줌.
- 결과를 list로 바꾸면 메모리 사용량이 늘어난다.
- `generator-like object`
    - 메모리에 즉시 저장하지 않음.
    - 코드만 메모리에 가지고 있다.
    - 실행하면, 그 이후 메모리를 차지하게 된다.
    - 한 번 출력된 요소/결과, 메모리에서 바로 제거된다.
    - 주소만 출력되면 GLO라 생각하면 된다.
    - `map()`

```python
a = zip('abc', list('def')); a
```

```python
type(a)
```

```python
b = list(zip('abc', list('def'))); b
```

## N-gram 만들기

```python
import re
words = re.sub(r'[(][\w\W]+?[)]', '',
                open('./07_data/07_Obama.txt', 'r').read().lower()).split()
words = [re.sub(r'^\W+|\W+$', '', w) for w in words]
words = [w for w in words if len(w) > 0]; len(words)
len(words)
```

```python
words[:4]
```

```python
bi_gram = list(zip(words[0:(len(words)-1)],
                    words[1:len(words)]))
```

```python
bi_gram[:5]
```

```python
bi_gram[-5:]
```

```python
tri_gram = list(zip(words[0:(len(words)-2)],
                    words[1:(len(words)-1)],
                    words[2:len(words)]))
```

```python
tri_gram[:5]
```

```python
tri_gram[-5:]
```

## Bi-Gram 빈도 DataFrame
- four-gram부터는 텍스트 분량이 상당히 커야 유의미하다.

```python
from collections import Counter
import pandas as pd
bi_gram = [' '.join(i) for i in bi_gram]
DF = pd.DataFrame({'Freq':Counter(bi_gram)}
                    ).sort_values(by='Freq', ascending=False)
DF.head()
```

`uni-gram`으로 봤을 때보다 훨씬 의미가 잘 보인다.
